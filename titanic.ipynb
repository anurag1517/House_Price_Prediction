```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.compose import ColumnTransformer

numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

```

```python
df_without_label=train_df.drop('Survived',axis=1)
df_without_label.head()
```

```python
label=train_df['Survived']
label.head()
```

```python
df_without_label.info()
```

```python
for col in ['Name','Sex','Ticket','Embarked']:
  df_without_label[col]=LabelEncoder().fit_transform(df_without_label[col] )
df_without_label.head()
```

```python
numeric_df_apipe=numeric_transformer.fit_transform(df_without_label)
numeric_df_apipe
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(numeric_df_apipe,label)
```

```python
some_data=df_without_label.iloc[:10]
some_labels=label.iloc[:10]
some_data_prepared=numeric_transformer.transform(some_data)
predicted=model.predict(some_data_prepared)
print('Predictions:', model.predict(some_data_prepared))
print('Labels:', list(some_labels))
```

```python
from sklearn.metrics import log_loss
loss=log_loss(some_labels,predicted)
print(loss)
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(some_labels,predicted)
print(accuracy)
```

```python
from sklearn.pipeline import Pipeline
from sklearn import preprocessing
from sklearn.impute import SimpleImputer
my_pipeline=Pipeline([
    ('imputer',SimpleImputer(strategy='median')),
    ('std_scaler',preprocessing.StandardScaler())
])
```

```python
X=housing_df.drop('MEDV',axis=1)
Y=housing_df['MEDV'];
X_prepared=my_pipeline.fit_transform(X)
```

```python
X_prepared_df=pd.DataFrame(X_prepared,columns=X.columns)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X_prepared,Y,test_size=0.2,random_state=42)
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
split=StratifiedShuffleSplit(n_splits=2,test_size=0.2,random_state=42)
for train_index,test_index in split.split(X_prepared_df,X_prepared_df['CHAS']):
  strat_train_set=X_prepared_df.loc[train_index]
  strat_test_set=X_prepared_df.loc[test_index]
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,Y_train)
```

```python
from sklearn.metrics import mean_squared_error
housing_predictions=model.predict(X_train)
mse=mean_squared_error(Y_train,housing_predictions)
rmse=np.sqrt(mse)
housing_prediction_test=model.predict(X_test)
mse_test=mean_squared_error(Y_test,housing_prediction_test)
rmse_test=np.sqrt(mse_test)
print(rmse)
print(rmse_test)
```

```python
from matplotlib import pyplot as plt
plt.scatter(Y_train,housing_predictions)
plt.xlabel('Actual')
plt.ylabel('Predicted')
```

```python
plt.scatter(Y_test,housing_prediction_test)
plt.xlabel('Actual')
plt.ylabel('Predicted')
```
